1. (3 point) Display pokemons by pokedex number, English name, and the number of its possible moves. Sort pokemons by the number of possible moves (from largest to smallest). Resultant dataframe expected by this question is shown in `pandas4_question1_answer.csv`.
  - The number of possible moves regarding a pokemon is considered from their type1 and type2.
  - Be careful not to over count the possible moves when type1 and type2 are the same.
  - Be careful not to count the same move more than once.

```python
path_move = '/content/drive/MyDrive/DADS5001 อ.ฐิติรัตน์ อ.เอกรัฐ/pokemon-move-data.csv'
path_pkm = '/content/drive/MyDrive/DADS5001 อ.ฐิติรัตน์ อ.เอกรัฐ/pokemon.csv'
move_df = pd.read_csv(path_move)
display(move_df.head())
pkm_df = pd.read_csv(path_pkm)
display(pkm_df.head())
display(pkm_df.info())
```

```python
move_df['Type'] = move_df['Type'].apply(lambda x : x.lower())

df1 = pd.merge(pkm_df[['pokedex_number', 'name', 'type1','type2']],
                       move_df[['Name', 'Type']],left_on= 'type1',right_on='Type')
df2 = pd.merge(pkm_df[['pokedex_number', 'name', 'type1','type2']],
               move_df[['Name', 'Type']], left_on= 'type2', right_on='Type')

df = pd.concat([df1, df2])
df = df[~df[['pokedex_number','name','Name']].duplicated()]
df = df.groupby(['pokedex_number','name']).count().reset_index()
df = df[['pokedex_number','name','Name']].rename(columns = {'Name': 'number of possible moves'})
df = df.sort_values(by = df.columns[-1], ascending= False)
df = df.reset_index(drop = True)
df
```
