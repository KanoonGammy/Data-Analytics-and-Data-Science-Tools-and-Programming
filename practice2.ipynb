1. (1 point) Based on the 'type1' attribute, conclude the numbers of each pokemon type. Sort the results descendingly based on the numbers of pokemons.

2. (1 point) Display all types of pokemons (considering 'type1' only) from the best to the worst stats.

- First, consider 'hp', the larger the better. Then, consider the average between 'attack' and 'defense', the larger the better
- The final display should include at least 'type1', 'hp', 'attack' and 'defense' columns.

3. (1.5 point) Display all pokemons whose 'hp' is greater than the average hp value computed from all pokemons with the same 'type1'. Sort the result by 'type1' (ascendingly) then by 'hp' (descendingly).

```python
"1: Based on the 'type1' attribute, conclude the numbers of each pokemon type. Sort the results descendingly based on the numbers of pokemons."
pkm = pkm_data.copy()
pkm.groupby('type1').size().sort_values(ascending=False).reset_index().rename(columns={0: 'numbers'})
```

```python
"2: Display all types of pokemons (considering 'type1' only) from the best to the worst stats."
pkm = pkm_data.copy()
drop_cols = [col for col in pkm.columns if pkm[col].dtype == 'object' and col != 'type1']
# print(drop_cols)
"First, consider 'hp', the larger the better. Then, consider the average between 'attack' and 'defense', the larger the better."
"The final display should include at least 'type1', 'hp', 'attack' and 'defense' columns."

pkm['attack+defense'] = (pkm['attack']+ pkm['defense'])/2
display(pkm.groupby('type1').agg({'hp':'mean', 'attack':'mean', 'defense': 'mean', 'attack+defense':'mean'}).sort_values(['hp','attack+defense'], ascending = [False,False]))
```

```python
"3: Display all pokemons whose 'hp' is greater than the average hp value computed from all pokemons with the same 'type1'. Sort the result by 'type1' (ascendingly) then by 'hp' (descendingly)."
pkm = pkm_data.copy()
pkm['hp_mean'] = pkm.groupby('type1')['hp'].transform('mean')
display(pkm[pkm['hp'] > pkm['hp_mean']].sort_values(['type1','hp'], ascending = [True,False])[['name','pokedex_number','type1','hp','attack','defense']].set_index('name'))
```
